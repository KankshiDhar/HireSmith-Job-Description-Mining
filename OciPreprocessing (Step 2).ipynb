```python
#Creating a key to merge on InterviewDate + Job Title for pr_oci
df3['key1'] = df3["Date"] + df3["Job Title"] + df3["Employer"]
```

```python
#Creating a key to merge on InterviewDate + Job Title + Company Name for oci_final
merged_oci['key1'] = merged_oci["Interview Date"] + merged_oci["Job Title"] + merged_oci["Company Name"]
```

```python
#rev = df1.merge(df2,on=['key1'],how='outer') , 
oci = pd.merge(merged_oci , df3[['key1','UG Opps', 'MS Opps', 'MBA Opps']], on = 'key1', how ='left')
oci.head()
```

```python
oci = oci[oci['Job Description'].notna()]
```

```python
oci = oci[oci['Interview Date'].notna()]
```

```python
oci['Industry']=oci['Industry'].fillna("Other")
```

```python
oci = oci.reset_index(drop =True)
```

```python
#Splitting the data on the basis of month, year and day
mt=[]
dy=[]
year=[]
for i in range(len(oci)):
    date=oci.loc[i,'Interview Date'].split('/')
    date1=date[2].split(' ')
    mt.append(date[0])
    dy.append(date[1])
    year.append(date1[0])
        
oci['month']=mt
oci['day']=dy
oci['year']=year
```

```python
oci=oci.sort_values('year')
```

```python
oci = oci.reset_index(drop = True)
```

```python
oci.head()
```

```python
#####################  Industry  ##################################
```

```python
#### Splitting multiple industries into rows
ind = []
result = pd.DataFrame()
for i in range(len(oci)):
        ind = oci.loc[i,'Industry'].split(',')
        if len(ind) > 1:
            for j in range(len(ind)):
                result = result.append(oci.loc[[i]].assign(Industry = ind[j]), ignore_index=True)
        else:
            result = result.append(oci.loc[[i]], ignore_index=True)
```

```python
result = result.reset_index(drop = True)
```
